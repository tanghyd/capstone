```python
%matplotlib inline
%load_ext autoreload
%autoreload 2

from capstone import load_spacy_model, load_metadata, match_triggers
```

```python
# build nlp model
nlp = load_spacy_model(output_type='doc') 

# build capstone metadata from source
metadata = load_metadata()
```

```python
# subset list of filenames to load based on subset data
# subset_filenames = [file.name for file in subset_reports_path.glob('*.json')]
# filenames = [filename for filename in metadata.filename.values if filename in subset_filenames]

# specify filenames to load
filenames = metadata.filename.tolist()[:1000]  # loading first 100 files

# extract triggers per sentence - takes about 5 minutes on my PC to load 33805 reports
#file_triggers = match_triggers(filenames, nlp=nlp, save_json=True, json_file='sample_triggers.json')
file_triggers = match_triggers(filenames, nlp=nlp, save_json=False)

# merge with metadata
df = file_triggers.merge(metadata, on='filename')
```
