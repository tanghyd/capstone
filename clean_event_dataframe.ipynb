```python
from pipeline.data.metadata import get_report_data
import pandas as pd

df = pd.read_csv('data/predictions/event_chunks-d2v_v50-c3-e20.csv')
cols = ['event_id', 'anumber', 'report_type','filename', 'sentence_count', 'counts', 'lower_idx',
       'upper_idx', 'length', 'prop','chars', 'prob', 'pred', 'tokens']

capstone_files, files = get_report_data(return_files=True)

event_text = [' '.join(files[event.filename][event.lower_idx : event.upper_idx + 1]).strip() for event in df.itertuples()]

cols += ['event_text']

df['event_text'] = event_text
```

```python
df[cols].sort_values('prob', ascending=False)
```

```python
df.to_csv('data/predictions/near_miss_events-sorted_by_prob.zip', index_label='index', compression='zip')
```
